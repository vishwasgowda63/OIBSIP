```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset task1/menu.csv")

df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.columns
```

```python
df.groupby("Category")["Calories"].mean().plot(kind="bar")

plt.title("Average Calories by Category")
plt.ylabel("Calories")
plt.xticks(rotation=45)

plt.show()
```

```python
top_items = df.sort_values(by="Calories", ascending=False).head(10)

plt.figure(figsize=(10,5))

sns.barplot(x="Calories", y="Item", data=top_items)

plt.title("Top 10 High Calorie Items")

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Protein"], bins=20)

plt.title("Protein Distribution")

plt.show()
```

```python
df.groupby("Category")["Sodium"].mean().plot(kind="bar")

plt.title("Average Sodium by Category")

plt.ylabel("Sodium")

plt.xticks(rotation=45)

plt.show()
```

# Insights

1. Some food categories contain very high calories.
2. Sodium levels are high in certain menu items.
3. High calorie foods usually contain more fat.
4. Protein distribution varies across menu items.
5. Heatmap shows strong relationships between calories, fat, and cholesterol.

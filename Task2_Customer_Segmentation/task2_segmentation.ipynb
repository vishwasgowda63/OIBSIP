```python
df = df.drop_duplicates()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Income"], bins=30)

plt.title("Income Distribution")

plt.savefig("income_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=20)

plt.title("Age Distribution")

plt.savefig("age_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="Income", y="MntTotal", data=df)

plt.title("Income vs Total Spending")

plt.savefig("income_vs_spending.png")

plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.savefig("heatmap_task2.png")

plt.show()
```

```python
spending = df[[
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts'
]].sum()

spending.plot(kind='bar')

plt.title("Top Spending Categories")

plt.savefig("top_spending_categories.png")

plt.show()
```

# Insights

1. Customers with higher income tend to spend more.
2. Wine and meat products have higher sales compared to other products.
3. Age distribution shows most customers are middle-aged.
4. Heatmap reveals strong relationships between income and spending.
5. Customer purchasing patterns can help businesses target marketing strategies.
